```python
#Importing Pandas and reading data from file 'Motor_Vehicle_Collisions_Vehicles.csv'
import pandas as pd
path = "/content/drive/MyDrive/Motor_Vehicle_Collisions_Vehicles.csv"
df = pd.read_csv(path)
```

```python
#Extract year-wise data
df['CRASH_DATE'] = pd.to_datetime(df['CRASH_DATE'])
#finding rows between dates
start_date = '11/01/2017'
end_date   = '10/31/2019'
mvc_data = df.query('CRASH_DATE >= @start_date and CRASH_DATE <= @end_date')
#writing data to a Vehicle CSV file
mvc_data.to_csv('/content/drive/MyDrive/Vehicle.csv')
#reading the Vehicle CSV file
year_wise = pd.read_csv("/content/drive/MyDrive/Vehicle.csv")
```

```python
#method returns the first 5 rows if a number is not specified.
year_wise.head()
```

```python
#extracting year and month
year_wise['CRASH_DATE'] = pd.to_datetime(year_wise['CRASH_DATE'])
year_wise['YEAR'], year_wise['MONTH'] = year_wise['CRASH_DATE'].dt.year, year_wise['CRASH_DATE'].dt.month
```

```python
# first 5 rows of a dataframe
year_wise.head()
```

```python
#Dropping rows of the VEHICLE_MAKE column which contains null values
year_wise=year_wise.dropna(subset=['VEHICLE_MAKE'])
```

```python
year_wise
```

```python
#replaces the NULL values with a 0.
year_wise=year_wise.fillna(0)
```

```python
year_wise
```

```python
import re
# Function to clean the names
def Clean_names(vehicle_name):
#Regex to search for hyphen(-) in the name followed by any #characters repeated any number of times
    if re.search('\-.*', vehicle_name):
        # Extract the position of beginning of pattern
        pos = re.search('\-.*', vehicle_name).start()
        # return the cleaned name
        return vehicle_name[:pos]
    else:
        # if clean up needed return the same name
        return vehicle_name

# Update the vehicle_make columns
year_wise['VEHICLE_MAKE'] = year_wise['VEHICLE_MAKE'].apply(Clean_names)
```

```python
#Extracting specific Vehicle makes from the VEHICLE_MAKE column
filtered_data=year_wise[year_wise['VEHICLE_MAKE'].str.contains('(TOYT|VOLK|ISU|YAMA)', regex = True)]
```

```python
#Replacing old values with new values
filtered_data['VEHICLE_MAKE'] = filtered_data['VEHICLE_MAKE'].replace(['TOYT ','TOYTA','TOYT TACOMA'],'TOYT')
```

```python
filtered_data['VEHICLE_MAKE'] = filtered_data['VEHICLE_MAKE'].replace(['ISU ','ISUZU','ISUZU TRUCK','ISUZU XXX','ISUZ','ISU 2018 VAN'],'ISU')
```

```python
filtered_data['VEHICLE_MAKE'] = filtered_data['VEHICLE_MAKE'].replace(['YAMA ','YAMAHA','ATV YAMAHA','YAMAH'],'YAMA')
```

```python
filtered_data['VEHICLE_MAKE'] = filtered_data['VEHICLE_MAKE'].replace(['VOLKS','VOLK '],'VOLK')
```
